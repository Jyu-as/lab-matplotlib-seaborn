#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
import math
```

```python
math.e
```

```python
y1 = x**2
y2 = (math.e)**x
```

```python
# your code here

fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,5))
fig.suptitle('Linear & Log scale', fontsize = 18,y=1.1)
ax[0].plot(x,y1,lw=5, c='r', ls='--')
ax[0].set_title('Linear scale')
ax[0].set_xscale('linear')
ax[0].set_yscale('linear')

ax[1].plot(x,y2,lw=5, c='b', ls='-.')
ax[1].set_title('Log scale')
ax[1].set_xscale('log')
ax[1].set_yscale('log')
plt.tight_layout()
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(10,5))
fig.suptitle('Linear & Log scale', fontsize = 18,y=1.1)
ax[0].plot(x,y1,lw=5, c='r', ls='--',label='y=x**2')
ax[0].set_title('Linear scale')
ax[0].set_xscale('linear')
ax[0].set_yscale('linear')

ax[1].plot(x,y2,lw=5, c='b', ls='-.',label='y=e**x')
ax[1].set_title('Log scale')
ax[1].set_xscale('log')
ax[1].set_yscale('log')
for axes in ax:
    axes.legend()
plt.tight_layout()
plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
# your code here
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(x=titanic['Fare'],y=titanic['Age'])
ax.set_xlim([0,600])
ax.set_ylim([0,90])
ax.set_ylabel("Age")
ax.set_xlabel("Fare")
plt.show()
```

```python
# Method 2 - seaborn
# your code here
fig, ax = plt.subplots(figsize=(12,8))
sns.regplot(x=titanic['Fare'],y=titanic['Age'],data=titanic)
ax.set_xlim([0,600])
ax.set_ylim([0,90])
ax.set_ylabel("Age")
ax.set_xlabel("Fare")
plt.show()
```
